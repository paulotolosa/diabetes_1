```python
import seaborn as sns
import matplotlib.pyplot as plt

# Establecer el estilo de Seaborn
sns.set(style="darkgrid")

# Crear el histograma con hue
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Age', hue='Outcome', bins=30, multiple="stack", palette='dark')

# Agregar etiquetas y título
plt.title('Histograma con Hue: Age vs Outcome')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')

# Mostrar el histograma
plt.legend(title='Outcome', labels=['Negativo', 'Positivo'])
plt.show()
```

```python
# Establecer el estilo de Seaborn
sns.set(style="whitegrid")

# Crear una figura y ejes para los subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Boxplot: Glucose vs Outcome
sns.boxplot(data=df, x='Outcome', y='Glucose', hue='Outcome', palette='Set2', ax=axes[0], legend=False)
axes[0].set_title('Glucose vs Outcome')
axes[0].set_xlabel('Outcome')
axes[0].set_ylabel('Glucose')

# Boxplot: BMI vs Outcome
sns.boxplot(data=df, x='Outcome', y='BMI', hue='Outcome', palette='Set2', ax=axes[1], legend=False)
axes[1].set_title('BMI vs Outcome')
axes[1].set_xlabel('Outcome')
axes[1].set_ylabel('BMI')

# Ajustar el espacio entre los subplots
plt.tight_layout()

# Mostrar los boxplots
plt.show()

```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Crear la figura y los ejes 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatterplot 3D
scatter = ax.scatter(df['Glucose'], df['BMI'], df['Insulin'], c=df['Outcome'], cmap='coolwarm', s=50)

# Agregar etiquetas y título
ax.set_xlabel('Glucose')
ax.set_ylabel('BMI')
ax.set_zlabel('Insulin')
ax.set_title('Scatterplot 3D: Glucose, BMI, Insulin')

# Agregar leyenda
legend1 = ax.legend(*scatter.legend_elements(), title="Outcome")
ax.add_artist(legend1)

# Mostrar el scatterplot 3D
plt.show()

```
